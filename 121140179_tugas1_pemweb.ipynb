```python
%%html
<html lang="en">
    <!--head-->
    <head>
    <title>Bahasa Pemrograman C++</title>
    <script src="assets\style\slider.js"></script>
    <link rel="stylesheet" href="assets\style\style.css">
    </head>

    <!--body-->
    <body>
        <!--header-->
        <header>
            <div class="jumbotron">
                <h1 id="awal">Bahasa Pemrograman C++</h1>
            </div>
            <nav>
                <ul>
                    <li class="box-li-nav"><a href="#definisi">Definisi</a></li>
                    <li class="box-li-nav"><a href="#sejarah">Sejarah</a></li>
                    <li class="box-li-nav"><a href="#variabel">Variabel</a></li>
                    <li class="box-li-nav"><a href="#keunggulan">Keunggulan</a></li>
                    <li class="box-li-nav"><a href="#contoh">Contoh</a></li>
                    <li class="box-li-nav"><a href="#kesimpulan">Kesimpulan</a></li>
                    <li class="scroll-up"><a href="#awal">&uparrow;</a></li>
                </ul>
            </nav>
        </header>

        <!--main-->
        <main>
            <div id="content">
                <article  id="definisi" class="card">
                    <section>
                        <h3>Definisi bahasa c++</h3>
                        <img src="assets/image/ngoding.jpg" class="featured-image" alt="pemrograman">
                        <p>Bahasa pemrograman c++ adalah bahasa pemrograman yang dikembangkan dari bahasa c, diciptakan oleh Bjarne Stroustrup,
                            dan dikembangkan di Bell Labs (Dennis Ritchie). Saat ini, bahasa c++ digunakan untuk membuat berbagai aplikasi yang bermanfaat, menarik,
                            dan menghibur seperti aplikasi editing gambar, perangkat lunak gawai, video game, dan bahkan sistem operasi modern.
                            Sintaks penulisan c dan c++ hampir mirip, tetapi berbeda dalam langkah penyelesaian masalah.
                        </p>
                </section>
            </article>
            
            <article id="sejarah" class="card">
                <section>
                    <h3>Sejarah bahasa c++</h3>
                    <p>Pada awalnya, bahasa tersebut dirancang sebagai bahasa pemrograman yang dijalankan pada sistem Unix. Pada perkembangannya, 
                        versi ANSI (American National Standards Institute) pada bahasa pemrograman C menjadi versi dominan, meskipun versi tersebut 
                        sekarang jarang dipakai dalam pengembangan sistem dan jaringan maupun untuk sistem embedded.
                    </p>
                    
                    <p>Bjarne Stroustrup pada Bell Labs pertama kali mengembangkan C++ pada awal 1980-an. Untuk mendukung fitur-fitur pada C++, 
                        dibangun efisiensi dan sistem support untuk pemrograman tingkat rendah (low level coding). Pada C++ ditambahkan konsep-konsep 
                        baru seperti class dengan sifat-sifatnya seperti inheritance dan overloading. Salah satu perbedaan yang paling mendasar 
                        dengan bahasa C adalah dukungan terhadap konsep pemrograman berorientasi objek (object-oriented programming).
                    </p>
                </section>
            </article>

            <article id="variabel" class="card">
                <h3>Variabel</h3>
                <p>Variabel adalah penanda identitas yang digunakan untuk menampung suatu nilai. Artinya, variabel akan menunjukkan suatu lokasi yang ada di memori komputer atau RAM. Jadi, saat Anda membuat satu variabel, akan ada satu slot memori untuk menampung nilai tersebut.</p>
                <p>Bahasa C++ termasuk bahasa pemrograman dengan konsep strongly typed. Artinya, saat melakukan deklarasi, Anda harus memberi tipe data di dalam setiap variabel.</p>
                <p>Berikut ini adalah tipe-tipe data yang ada di dalam variabel C++:</p>
                <ul>
                        <li>bool: tipe data berupa nilai boolean, yaitu True or False</li>
                        <li>char: tipe data huruf dari A sampai Z</li>
                        <li>int: tipe data berupa angka</li>
                        <li>float dan double: tipe data berupa angka pecahan, contohnya 0.5</li>
                        <li>string: tipe data berupa kumpulan karakter seperti “Hello World!”</li>
                    </ul>
            </article>
            
            <article id="keunggulan" class="card">
                <h3>Keunggulan Bahasa Pemrograman C++</h3>
                <h4>1. Portabel</h4>
                <p>Bahasa pemrograman C++ itu portabel. Artinya, Anda bisa melakukan coding, lalu melakukan compile pada sistem operasi yang berbeda-beda. Misalnya, Anda bisa menulis kode di sistem operasi Linux dan melakukan compile kode tersebut di Windows. Dengan kata lain, saat belajar pemrograman c++, Anda bebas menggunakan sistem operasi apapun.</p>
                
                <h4>2. Object Oriented</h4>
                <p>C++ adalah bahasa pemrograman yang object oriented atau berfokus pada objek. Artinya, data dan function disatukan dalam kelas dan objek untuk bekerjasama memecahkan sebuah masalah. Objek yang saling terkait akan mempercepat penulisan kode. Sebab, Anda bisa menggunakan kode class yang sama melalui fitur inheritance untuk menghindari pengulangan kode. OOP juga memudahkan Anda untuk mendeteksi permasalahan apabila ada kendala pada aplikasi.</p>
                
                <h4>3. Mudah Mengatur Memori</h4>
                <p>Bahasa pemrograman C++ memiliki Dynamic Memory Allocation atau kemampuan mengatur memori dengan dinamis. Dengan pengaturan tersebut, Anda lebih mudah mendapatkan memori tambahan saat program berjalan. Kemudahan ini cukup membantu programmer untuk mengatur Linked List atau Tree sesuai kebutuhan.</p>
                
                <h4>4. Memiliki Skalabilitas yang Baik</h4>
                <p>C++ memiliki skalabilitas yang cukup baik. Artinya, program yang dibuat bisa dikembangkan sesuai kebutuhan Anda. Jadi, Anda bisa membuat program sederhana yang kecil dulu sebelum mengembangkannya menjadi aplikasi yang lebih besar.</p>

                <h4>5. Eksekusi Lebih Cepat</h4>
                <p>Bahasa pemrograman C++ adalah bahasa yang memungkinkan Anda melakukan low level manipulation. Artinya, meskipun merupakan sebuah high level language, C++ mampu membuat eksekusi kode bisa berjalan lebih cepat di komputer karena minim jeda waktu saat interpretasi kode yang dilakukan.</p>
            </article>
            
            <article id="contoh" class="card">
                <h3>Contoh Program  C++ yang Sering Digunakan </h3>
                <p>Setelah mempelajari konsep dasar bahasa pemrograman C++, Anda bisa menggunakannya untuk berbagai pengembangan aplikasi. Berikut ini adalah contoh aplikasi yang menggunakan bahasa pemrograman C++:</p>
                <h4>Aplikasi Grafis</h4>
                <p>Coding C++ banyak digunakan untuk mengembangkan aplikasi pengolah grafis. Misalnya, Adobe Photoshop Adobe Illustrator, Adobe dan Premiere. Sebagian aplikasi ini membutuhkan proses image processing yang cukup intensif dan C++ dianggap cukup sesuai untuk membangun program tersebut.</p>
                <p>Nah, jika belajar pemrograman c++, bukanlah suatu hal yang mustahil bila di masa mendatang Anda bisa menciptakan software kompleks seperti yang kami sebutkan di atas.</p>
                
                <h4>Software Games</h4>
                <p>Bahasa pemrograman C++ juga digunakan untuk mengembangkan software games 3D yang rumit. Salah satu game terkenal yang dikembangkan dengan C++ dan bahkan sering Anda mainkan sekarang adalah Valorant.</p>
                <p>C++ cocok untuk pengembangan game karena kebutuhan proses eksekusi yang cepat. Nah, dengan kemampuan low level manipulation, C++ tentu mampu melakukannya lebih baik dibanding bahasa pemrograman lain.</p>
                
                <h4>Browser</h4>
                <p>Cukup banyak browser terkenal yang dikembangkan dengan bahasa pemrograman C++ seperti Google Chrome dan Mozilla Firefox. Salah satu alasannya tentu adalah portability yang dimiliki oleh C++. Sebab, sebuah browser tentu akan dibuat agar multiplatform untuk memudahkan penggunanya.</p>
                <p>Pun demikian, sebuah browser biasanya tidak hanya dikembangkan dengan bahasa pemrograman C++ saja, tapi juga dengan bahasa lain seperti Java dan lainnya.</p>
                
                <h4>Sistem Operasi</h4>
                <p>Sebagian besar sistem operasi dikembangkan dengan C++, seperti Windows dan Mac OS. Namun tentu saja dengan didukung oleh bahasa pemrograman lain, misalnya Swift.</p>
                <p>Salah satu alasan utama penggunaan coding C++ pada sistem operasi adalah kecepatan eksekusi di komputer yang menjadi kunci sebuah OS dapat berjalan dengan baik.</p>
            </article>
            
            <article id="kesimpulan" class="card">
                <h3>Kesimpulan</h3>
                <p>Belajar bahasa pemrograman C++ dengan baik memungkinkan Anda untuk dapat mengembangkan berbagai macam program dengan mudah. Mulai dari aplikasi, games, database hingga media player.</p>
                <p>Mempelajari bahasa pemrograman C++ tidak terlalu sulit jika Anda sudah tahu konsep dasarnya, meliputi variabel, control structure, data structure, syntax. Oh iya, jangan lupa untuk menguasai tools yang diperlukan untuk belajar C++.</p>
                <p>Dengan berbagai keunggulan yang dimiliki seperti portability, OOP, memori dinamis, dan eksekusi yang cepat, C++ bisa mewujudkan impian Anda untuk membuat aplikasi yang bermanfaat. Hal ini tentu akan membuat Anda semakin semangat belajar untuk menjadi seorang developer yang andal.</p>
            </article>
            
            <div class="slides">
                <div class="slide">
                  <img src="assets/image/iklan 5.jpg" alt="iklan 5">
                </div>
                <div class="slide">
                  <img src="assets/image/iklan 4.jpg" alt="iklan 4">
                </div>
                <div class="slide">
                  <img src="assets/image/iklan 3.jpg" alt="iklan 3">
                </div>
                <div class="slide">
                  <img src="assets/image/iklan 2.jpg" alt="iklan 2">
                </div>
                <div class="slide">
                  <img src="assets/image/iklan 1.jpg" alt="iklan 1">
                </div>
                <div class="navigation">
                  <a class = "prev" onclick = "plusSlides(-1)">&#10094;</a>
                  <a class = "next" onclick = "plusSlides(-1)">&#10095;</a>
                </div>
              </div>
            </div>

            <aside id="informasi lain" class="aside-card">
                <article>
                    <header>
                        <h3 class="h3-element-aside">Bahasa C++</h3>
                        <figure>
                            <img src="assets/image/images.png" lenght="100" width="100" alt="logo c++" class="logo-element-aside">
                            <figcaption class="h3-element-aside">Logo C++</figcaption>
                        </figure>    
                    </header>
                    <section>
                        <h3 class="h3-element-aside">Informasi Lainnya</h3>
                        <table>
                           <tr>
                              <th>Keluarga</th>
                              <td>C</td>
                            </tr>
                           <tr>
                              <th>Perancang</th>
                              <td>Bjarne Stroustrup</td>
                            </tr>
                            <tr>
                                <th>Rilis Perdana</th>
                                <td>1985</td>
                            </tr>
                            <tr>
                                <th>Tipe Sistem</th>
                                <td>Statis, nominal, sebagian interfrensi</td>
                            </tr>
                            <tr>
                                <th>Paradigma</th>
                                <td>Multi-paradigma: prosedural, fungsional, orientasi objek, generik, modular</td>
                            </tr>
                        </table>
                    </section>
                </article>
            </aside>
            
            <aside class="aside-card">
                <section>
                    <h3 class="h3-element-aside">Pembuat Website</h3>
                    <img src="assets/image/WhatsApp Image 2022-05-26 at 9.42.10 AM.jpeg" alt="foto profil" class="profile-photo-aside">
                    <table>
                        <tr>
                           <th>Nama</th>
                           <td>Farhan Apri Kesuma</td>
                         </tr>
                        <tr>
                           <th>TTL</th>
                           <td>Pringsewu, 4 Januari 2004</td>
                         </tr>
                         <tr>
                             <th>Usia</th>
                             <td>18 Tahun</td>
                         </tr>
                         <tr>
                             <th>Pekerjaan</th>
                             <td>Mahasiswa</td>
                         </tr>
                         <tr>
                             <th>Universitas</th>
                             <td>Institut Teknologi Sumatera</td>
                         </tr>
                         <tr>
                            <th>Prodi</th>
                            <td>Teknik Informatika</td>
                        </tr>
                         <tr>
                            <th>Hobi</th>
                            <td>Bermain Game</td>
                        </tr>
                     </table>
                </section>
            </aside>

        </main>

        <!--footer-->
        <footer>
            <p>Belajar Dasar Pemrograman Web &copy; 2022, Dicoding Academy</p>
        </footer>
    </body>
</html>
```
