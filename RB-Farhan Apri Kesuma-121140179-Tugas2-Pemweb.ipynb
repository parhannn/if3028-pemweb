## Tambahan: Melempar Data ke Halaman Detail

Karena saya memiliki rasa penasaran yang tinggi, saya mencoba untuk melempar data dari halaman form dengan query string dan method POST ke halaman detail menggunakan sedikit `JavaScript`. Berikut script pada form:

```javascript
document.getElementById("formRegistrasi").addEventListener("submit", function (event) {
    event.preventDefault();

    const nama = document.getElementById("nama").value;
    const email = document.getElementById("email").value;
    const telepon = document.getElementById("telepon").value;
    const alamat = document.getElementById("alamat").value;

    const queryString = `nama=${encodeURIComponent(nama)}&email=${encodeURIComponent(email)}&telepon=${encodeURIComponent(telepon)}&alamat=${encodeURIComponent(alamat)}`;
    window.location.href = `detail.html?${queryString}`;    
});

```
sedikit penjelasan kode:
- `event.preventDefault()`: Untuk mencegah pengiriman form secara default.
- `const -data- = document.getElementById('-data-').value`: Untuk mengambil nilai dari form.
- `const queryString`: Variabel yang berisi query string yang nantinya akan dilakukan Redirect.
- `window.location.href`: Variabel yang berisi Redirect untuk mengarah ke halaman detail sekaligus membawa data dari form.

---

Dan berikut adalah script pada detail untuk menangkap data:

```javascript
const url = new URLSearchParams(window.location.search);

document.getElementById('nama').textContent = url.get('nama');
document.getElementById('email').textContent = url.get('email');
document.getElementById('telepon').textContent = url.get('telepon');
document.getElementById('alamat').textContent = url.get('alamat');

```
sedikit penjelasan kode:
- `const url`: Untuk wadah url dari query string yang dikirim.
- `document.getElementById('-data-').textContent = url.get('-data-')`: Untuk meletakkan masing-masing nilai dari query string berdasarkan url yang dikirim (*redirect*) yakni nama, email, telepon, dan alamat.
